```python
import pandas as pd

df = pd.read_csv('final_data.csv')
df.head()
```

```python
# Group columns by pollutant type and weather parameters

site_co_cols = [col for col in df.columns if 'co' in col.lower() and 'site' in col.lower()]
site_no_cols = [col for col in df.columns if '_no' in col.lower() and 'site' in col.lower() and '_no2' not in col.lower()]
site_no2_cols = [col for col in df.columns if 'no2' in col.lower() and 'site' in col.lower()]
site_so2_cols = [col for col in df.columns if 'so2' in col.lower() and 'site' in col.lower()]
site_pm10_cols = [col for col in df.columns if 'pm10' in col.lower() and 'site' in col.lower()]
site_pm25_cols = [col for col in df.columns if 'pm25' in col.lower() and 'site' in col.lower()]
site_nox_cols = [col for col in df.columns if 'nox' in col.lower() and 'site' in col.lower()]
site_nh3_cols = [col for col in df.columns if 'nh3' in col.lower() and 'site' in col.lower()]
site_solar_cols = [col for col in df.columns if 'solar' in col.lower() and 'site' in col.lower()]

# Weather parameters
weather_cols = ['u10', 'v10', 'tp', 'tcc', 't2m', 'sp', 'e', 'cbh']

# Create separate dataframes for each pollutant type and weather
df_co = df[['Timestamp'] + site_co_cols + weather_cols]
df_no = df[['Timestamp'] + site_no_cols + weather_cols]
df_no2 = df[['Timestamp'] + site_no2_cols + weather_cols]
df_so2 = df[['Timestamp'] + site_so2_cols + weather_cols]
df_pm10 = df[['Timestamp'] + site_pm10_cols + weather_cols]
df_pm25 = df[['Timestamp'] + site_pm25_cols + weather_cols]
df_nox = df[['Timestamp'] + site_nox_cols + weather_cols]
df_nh3 = df[['Timestamp'] + site_nh3_cols + weather_cols]
df_solar = df[['Timestamp'] + site_solar_cols + weather_cols]

# Print the shape of each dataframe to verify the split
print("Shape of CO data:", df_co.shape)
print("Shape of NO data:", df_no.shape)
print("Shape of NO2 data:", df_no2.shape)
print("Shape of SO2 data:", df_so2.shape)
print("Shape of PM10 data:", df_pm10.shape)
print("Shape of PM2.5 data:", df_pm25.shape)
print("Shape of NOx data:", df_nox.shape)
print("Shape of NH3 data:", df_nh3.shape)
print("Shape of Solar Radiation data:", df_solar.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Set up the plotting style
sns.set_theme()
sns.set_palette("husl")

# Function to calculate mean values for each pollutant and weather parameter
def get_mean_values(dfs, names, weather_cols):
    means = {}
    for df, name in zip(dfs, names):
        if len(df.columns) > 1:  # If there is data
            # Calculate mean across all pollutant stations
            pollutant_cols = [col for col in df.columns if col not in ['Timestamp'] + weather_cols]
            means[name] = df[pollutant_cols].mean(axis=1)
    
    # Add weather parameters
    for col in weather_cols:
        means[col] = df[col]  # Use the last df since weather cols are same in all
    
    return pd.DataFrame(means)

# Function to create box plots
def plot_comparison_boxplots(combined_df, weather_cols):
    # Separate pollutants and weather
    pollutants = combined_df.drop(columns=weather_cols)
    
    # Plot pollutants
    plt.figure(figsize=(15, 8))
    data_to_plot = [pollutants[col].dropna() for col in pollutants.columns]
    plt.boxplot(data_to_plot, labels=pollutants.columns, vert=False)
    plt.title('Box Plot Comparison of Different Pollutants')
    plt.xlabel('Value')
    plt.tight_layout()
    plt.show()

# Function to plot correlation matrix
def plot_correlation_matrix(combined_df):
    corr_matrix = combined_df.corr()
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Matrix Between Pollutants and Weather Parameters')
    plt.tight_layout()
    plt.show()

# List of dataframes and their names
pollutant_dfs = [df_co, df_no, df_no2, df_so2, df_pm10, df_pm25, df_nox, df_nh3, df_solar]
pollutant_names = ["CO", "NO", "NO2", "SO2", "PM10", "PM2.5", "NOx", "NH3", "Solar Radiation"]

# Get mean values including weather parameters
combined_data = get_mean_values(pollutant_dfs, pollutant_names, weather_cols)

# Create box plots  
plot_comparison_boxplots(combined_data, weather_cols)

# Create correlation matrix
plot_correlation_matrix(combined_data)

# Print basic statistics
print("\nBasic Statistics for Pollutants and Weather Parameters:")
print(combined_data.describe())

print("\nSkewness:")
print(combined_data.skew())

print("\nKurtosis:")
print(combined_data.kurtosis())
```
